## Import the necessary libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

### Split the data into Target and Inputs

```python
X = data.drop('prognosis', axis = 1)
y = data['prognosis']
```

### Encode the Prognosis column

```python
encoder = LabelEncoder()
encoder.fit(y)
Y = encoder.transform(y)
```

### Train Test Split

```python
x_train,x_test,y_train,y_test = train_test_split(X,Y, train_size=0.75, random_state=42)
```

## Model Building

```python
from sklearn.datasets import make_classification
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
# create a dictionary to store models

models = {
    "SVC": SVC(kernel='linear'),
    "RandomForest": RandomForestClassifier(n_estimators=100, random_state=42),
    "GradientBoosting": GradientBoostingClassifier(n_estimators=100, random_state=42),
    "KNeghbors": KNeighborsClassifier(n_neighbors=5),
    "MultinomialNB": MultinomialNB()
}
for model_name, model in models.items():
    # Train the model
    model.fit(x_train,y_train)

    # Test the model
    predictions = model.predict(x_test)

    # Calculate accuracy
    accuracy = accuracy_score(y_test,predictions)

    # Confusion matrix
    cm = confusion_matrix(y_test,predictions)

    print(f"{model_name} accuracy: {accuracy}")
    print(f"{model_name} confusion matrix:")
    print(np.array2string(cm,separator=','))
```

### Single model prediction

```python
# selecting svc
svc = SVC(kernel='linear')
svc.fit(x_train,y_train)
ypred = svc.predict(x_test)
accuracy_score(y_test,ypred)
```

```python
import ast
# Function to safely parse and clean diet lists
def parse_and_clean_diet(diet_string):
    try:
        # If the diet_string is empty or just "[]", return an empty list
        if diet_string in ["", "[]"]:
            return []
        # Safely evaluate string to list
        diet_list = ast.literal_eval(diet_string)
        # Remove the first element if it's a list with at least 2 elements
        return diet_list[1:] if isinstance(diet_list, list) and len(diet_list) > 1 else []
    except (ValueError, SyntaxError):
        # If any error occurs during parsing, return an empty list
        return []

# Apply the function to clean the "Diet" column
diets['Diet'] = diets['Diet'].apply(parse_and_clean_diet)
```
